# **Convolutional Neural Networks (CNN)**

---

## **1. Definition**

A Convolutional Neural Network (CNN) is a type of deep neural network specifically designed to process **grid-like structured data such as images**. CNNs automatically learn spatial hierarchies of features through convolution operations.

---

## **2. Why CNN is Needed**

Traditional neural networks treat images as **flat vectors**, which ignores spatial relationships between pixels.

Example:

```
224 × 224 × 3 = 150,528 features
```

CNN solves this problem using:

- local receptive fields
- parameter sharing
- hierarchical feature extraction

---

## **3. Basic Architecture of CNN**

```
Input Image
     ↓
Convolution Layer
     ↓
Activation Function
     ↓
Pooling Layer
     ↓
Convolution Layer
     ↓
Pooling Layer
     ↓
Flatten Layer
     ↓
Fully Connected Layer
     ↓
Output Layer
```

---

## **4. Convolution Operation**

### **Definition**

Convolution is the process of applying a **kernel (filter)** to an input image in order to extract features.

---

### **Mathematical Representation**

```
S(i,j) = (X * K)(i,j)
```

Expanded form:

```
S(i,j) = Σ Σ X(i+m, j+n) K(m,n)
```

---

## **5. Kernel / Filter**

A kernel is a **small matrix used to detect patterns**.

Example kernel:

```
[ -1 -1 -1 ]
[ -1  8 -1 ]
[ -1 -1 -1 ]
```

---

## **6. Feature Maps**

The output of a convolution operation is called a **feature map**.

Feature maps highlight:

- edges
- textures
- shapes
- object parts

---

## **7. Stride**

Stride determines how far the filter moves across the image.

Example:

```
Stride = 1
```

Filter moves one pixel at a time.

---

## **8. Padding**

Padding adds extra pixels around the border of the image.

Types:

```
Valid Padding
Same Padding
```

---

## **9. Output Size Formula**

```
Output Size = (N − F + 2P) / S + 1
```

Where:

```
N = input size
F = filter size
P = padding
S = stride
```

---

## **10. Activation Functions in CNN**

Activation functions introduce **non-linearity**.

Common activations:

- ReLU
- Leaky ReLU
- GELU
- Sigmoid
- Tanh

Formula:

```
ReLU(x) = max(0, x)
```

---

## **11. Pooling Layer**

Pooling reduces spatial dimensions.

Types:

- Max Pooling
- Average Pooling

Example:

```
[1 3]
[2 4]

Max Pool → 4
```

---

## **12. Flatten Layer**

Flatten converts a **2D feature map into a 1D vector**.

Example:

```
[1 2
 3 4]

→

[1 2 3 4]
```

---

## **13. Fully Connected Layer**

Fully connected layers perform final classification.

```
z = Wx + b
```

---

## **14. Softmax Layer**

Softmax converts outputs into probabilities.

```
Softmax(x_i) = e^(x_i) / Σ e^(x_j)
```

---

## **15. Backpropagation in CNN**

Training steps:

1. Forward propagation
2. Loss calculation
3. Backpropagation
4. Weight update

---

## **16. CNN Model Example (TensorFlow)**

```python
import tensorflow as tf
from tensorflow.keras import layers, models

model = models.Sequential()

model.add(layers.Conv2D(32,(3,3),activation='relu',input_shape=(28,28,1)))

model.add(layers.MaxPooling2D((2,2)))

model.add(layers.Conv2D(64,(3,3),activation='relu'))

model.add(layers.MaxPooling2D((2,2)))

model.add(layers.Flatten())

model.add(layers.Dense(64,activation='relu'))

model.add(layers.Dense(10,activation='softmax'))

model.compile(
optimizer='adam',
loss='categorical_crossentropy',
metrics=['accuracy']
)

model.summary()
```
