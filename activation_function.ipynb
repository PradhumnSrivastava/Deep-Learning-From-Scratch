# **Activation Functions in Deep Learning**

## **Definition**

Activation functions are mathematical functions applied to the output of a neuron in a neural network. They introduce **non-linearity** into the model, allowing neural networks to learn complex patterns and relationships in data.

Without activation functions, a neural network would behave like a **simple linear model**, regardless of how many layers it contains.

---

# Why Activation Functions are Important

Neural networks compute:

```
z = Wx + b
```

If no activation function is applied, the network remains a **linear transformation**, which limits its ability to learn complex relationships.

Activation functions solve this by introducing **non-linear behavior**.

---

# **Sigmoid Function**

---

## **Definition**

The Sigmoid function maps input values into a range between **0 and 1**. It is commonly used in **binary classification problems** because its output can be interpreted as a probability.

---

## **Formula**

```
σ(x) = 1 / (1 + e^(-x))
```

---

## **Characteristics**

| Property | Description |
|------|------|
| Output Range | 0 to 1 |
| Type | Non-linear |
| Common Use | Binary classification output layer |

---

## **Example Code**

```python
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

---

# **Tanh Function**

---

## **Definition**

The Tanh (Hyperbolic Tangent) function maps input values into the range **-1 to 1**. It is similar to the sigmoid function but is **zero-centered**, which helps optimization during training.

---

## **Formula**

```
tanh(x) = (e^x - e^(-x)) / (e^x + e^(-x))
```

---

## **Characteristics**

| Property | Description |
|------|------|
| Output Range | -1 to 1 |
| Type | Non-linear |
| Advantage | Zero-centered outputs |

---

## **Example Code**

```python
def tanh(x):
    return np.tanh(x)
```

---

# **ReLU (Rectified Linear Unit)**

---

## **Definition**

ReLU is one of the most widely used activation functions in deep learning. It outputs the input value if it is positive and zero otherwise.

---

## **Formula**

```
ReLU(x) = max(0, x)
```

---

## **Characteristics**

| Property | Description |
|------|------|
| Output Range | 0 to ∞ |
| Computational Cost | Very low |
| Common Use | Hidden layers in deep networks |

---

## **Example Code**

```python
def relu(x):
    return np.maximum(0, x)
```

---

# **Leaky ReLU**

---

## **Definition**

Leaky ReLU is a variation of ReLU designed to solve the **dying ReLU problem**. Instead of outputting zero for negative inputs, it allows a small slope.

---

## **Formula**

```
LeakyReLU(x) = x      if x > 0
LeakyReLU(x) = αx     if x ≤ 0
```

Where:

```
α = small constant (e.g., 0.01)
```

---

## **Example Code**

```python
def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)
```

---

# **Softmax Function**

---

## **Definition**

Softmax converts a vector of values into a **probability distribution** where the sum of all probabilities equals 1. It is commonly used in **multi-class classification problems**.

---

## **Formula**

```
Softmax(x_i) = e^(x_i) / Σ e^(x_j)
```

---

## **Characteristics**

| Property | Description |
|------|------|
| Output Range | 0 to 1 |
| Output Sum | 1 |
| Common Use | Multi-class classification output layer |

---

## **Example Code**

```python
def softmax(x):
    exp_values = np.exp(x - np.max(x))
    return exp_values / np.sum(exp_values)
```

---

# **ELU (Exponential Linear Unit)**

---

## **Definition**

ELU is an activation function designed to improve learning speed and reduce bias shifts by allowing negative outputs.

---

## **Formula**

```
ELU(x) = x                 if x > 0
ELU(x) = α(e^x − 1)        if x ≤ 0
```

---

## **Example Code**

```python
def elu(x, alpha=1.0):
    return np.where(x > 0, x, alpha * (np.exp(x) - 1))
```

---

# **Swish Function**

---

## **Definition**

Swish is a modern activation function discovered by researchers. It performs better than ReLU in many deep learning architectures.

---

## **Formula**

```
Swish(x) = x * sigmoid(x)
```

---

## **Example Code**

```python
def swish(x):
    return x / (1 + np.exp(-x))
```

---

# **Comparison of Activation Functions**

| Activation Function | Output Range | Common Use |
|------|------|------|
| Sigmoid | 0 to 1 | Binary classification |
| Tanh | -1 to 1 | Hidden layers |
| ReLU | 0 to ∞ | Deep neural networks |
| Leaky ReLU | -∞ to ∞ | Prevent dying ReLU |
| Softmax | 0 to 1 | Multi-class classification |
| ELU | -∞ to ∞ | Faster learning |
| Swish | -∞ to ∞ | Modern deep learning models |

---

# Summary

Activation functions introduce **non-linearity** into neural networks, enabling them to learn complex patterns in data. Different activation functions are used depending on the type of problem and the architecture of the neural network.

Common choices include:

- **ReLU for hidden layers**
- **Sigmoid for binary classification**
- **Softmax for multi-class classification**
