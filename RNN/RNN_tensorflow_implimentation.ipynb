## **Import Required Libraries**

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense
```

## **Prepare the Dataset**

```python
import numpy as np

X = np.random.rand(1000, 10, 5)   # 1000 samples, 10 timesteps, 5 features
y = np.random.rand(1000, 1)
```

## **Build the RNN Model**

```python
model = Sequential([
    
    SimpleRNN(
        units=64,
        activation='tanh',
        input_shape=(10,5)
    ),
    
    Dense(1)
])
```

## **Compile the Model**

```python
model.compile(
    optimizer='adam',
    loss='mse',
    metrics=['mae']
)
```

## **Train the Model**

```python
history = model.fit(
    X,
    y,
    epochs=20,
    batch_size=32,
    validation_split=0.2
)
```

## **Evaluate the Model**

```python
model.evaluate(X, y)
```

## **Make Predictions**

```python
predictions = model.predict(X)
```

## **Stacked RNN (Multiple RNN Layers)**

```python
model = Sequential([
    
    SimpleRNN(128, return_sequences=True, input_shape=(10,5)),
    
    SimpleRNN(64),
    
    Dense(1)
])  
```

## **RNN with Dropout (Regularization)**

```python
from tensorflow.keras.layers import Dropout

model = Sequential([
    
    SimpleRNN(128, return_sequences=True, input_shape=(10,5)),
    
    Dropout(0.3),
    
    SimpleRNN(64),
    
    Dense(1)
])
```

## **Save and Load Model**

```python
model.save("rnn_model.h5")

# To load the model later
from tensorflow.keras.models import load_model

model = load_model("rnn_model.h5")
```
