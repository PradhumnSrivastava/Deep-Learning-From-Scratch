# **Training Process of Recurrent Neural Network (RNN)**

Recurrent Neural Networks (RNNs) are designed to process **sequential data** such as text, time series, audio, and video. Unlike traditional neural networks, RNNs maintain a **hidden state (memory)** that carries information from previous time steps.

During training, the RNN processes the sequence step-by-step and updates its weights using **Backpropagation Through Time (BPTT)**.

---

# **1. Preparing the Input Sequence**

RNNs work on **sequence data**.

Example sentence:

"I love Krishna"

Tokenized representation:

```
I      love      Krishna
x1     x2        x3
```

Sequence representation:

```
X = [x1, x2, x3]
```

Input tensor shape required for RNN:

```
(batch_size, time_steps, features)
```

Example:

```
(32, 10, 5)
```

Meaning:

* 32 samples in one batch
* 10 time steps in each sequence
* 5 features per time step

---

# **2. Initialize the Hidden State**

RNN maintains a **hidden state** which stores information from previous time steps.

The initial hidden state is usually initialized as zeros.

Example:

```python
import torch

h0 = torch.zeros(1, batch_size, hidden_size)
```

Where:

* `batch_size` → number of samples
* `hidden_size` → number of neurons in the hidden layer

---

# **3. Forward Pass Through Time Steps**

At each time step, the RNN processes the current input and the previous hidden state.

Mathematical formulation:

```
h_t = tanh(Wx * x_t + Wh * h_(t-1) + b)
```

Where:

* `x_t` = input at time step t
* `h_(t-1)` = previous hidden state
* `Wx` = input weight matrix
* `Wh` = recurrent weight matrix
* `b` = bias

Output equation:

```
y_t = Wy * h_t
```

---

## Example Time Flow

```
Step 1
x1 + h0 → h1
```

```
Step 2
x2 + h1 → h2
```

```
Step 3
x3 + h2 → h3
```

The final hidden state is used to generate the output.

---

# **4. Loss Calculation**

After forward propagation, the predicted output is compared with the actual label.

Example loss function:

```
Loss = MSE(y_pred, y_true)
```

Example implementation:

```python
loss = criterion(output, y)
```

Common loss functions:

* Mean Squared Error (MSE)
* Cross Entropy Loss
* Binary Cross Entropy

---

# **5. Backpropagation Through Time (BPTT)**

RNNs are trained using a special form of backpropagation called:

**Backpropagation Through Time (BPTT)**

Gradients propagate backward through all time steps.

Example gradient flow:

```
Loss
 ↓
y3
 ↓
h3
 ↓
h2
 ↓
h1
```

The gradients are calculated for all shared weights across time.

---

# **6. Gradient Calculation**

Gradients of the loss with respect to weights are computed automatically.

Mathematically:

```
∂Loss / ∂W
```

Example implementation:

```python
loss.backward()
```

This calculates gradients using the automatic differentiation engine.

---

# **7. Updating the Weights**

After gradients are computed, the optimizer updates the weights.

Example:

```python
optimizer.step()
```

Common optimizers:

* SGD (Stochastic Gradient Descent)
* Adam
* RMSprop

---

# **8. Training Loop**

The complete training process runs inside a training loop.

Example:

```python
for epoch in range(epochs):

    optimizer.zero_grad()

    outputs = model(X_train)

    loss = criterion(outputs, y_train)

    loss.backward()

    optimizer.step()

    print(loss.item())
```

Steps happening inside the loop:

1. Reset gradients
2. Perform forward pass
3. Compute loss
4. Backpropagation
5. Update weights

---

# **9. Making Predictions**

After training, the model can make predictions.

Example:

```python
predictions = model(X_test)
```

---

# **Complete RNN Training Workflow**

```
1 Prepare sequence data
        ↓
2 Initialize hidden state
        ↓
3 Forward pass through time steps
        ↓
4 Compute loss
        ↓
5 Backpropagation Through Time
        ↓
6 Compute gradients
        ↓
7 Update weights
        ↓
8 Repeat for multiple epochs
```

# **Common Problems During RNN Training**

## **Vanishing Gradient Problem**

Gradients become extremely small during backpropagation.

Result:

```
The model fails to learn long-term dependencies.
```

---

## **Exploding Gradient Problem**

Gradients become extremely large and make training unstable.

Solution:

**Gradient clipping**

Example:

```python
torch.nn.utils.clip_grad_norm_(model.parameters(), 1)
```

## **Working**

## Working of a Recurrent Neural Network (RNN) — In Depth

A **Recurrent Neural Network (RNN)** is designed to process **sequential data**, where the order of inputs matters. Unlike a traditional neural network that treats each input independently, an RNN processes data **step by step through time** and maintains an internal memory called the **hidden state**. This memory allows the network to remember information from previous steps and use it when processing the current input.

In an RNN, the input sequence is represented as (x_1, x_2, x_3, ..., x_t), where each (x_t) is the input at time step (t). For example, in a weather prediction task, each input could be the temperature of a particular day. The RNN processes these inputs sequentially, updating its internal memory at each step.

At the core of an RNN is the **hidden state** (h_t), which represents the network’s memory at time step (t). This hidden state is calculated using two pieces of information: the **current input** (x_t) and the **previous hidden state** (h_{t-1}). The previous hidden state carries information about the past inputs that the model has already processed. The RNN combines these two components through a linear transformation followed by a non-linear activation function to produce the new hidden state.

Mathematically, the hidden state update is defined as:

[
h_t = \tanh(W_x x_t + W_h h_{t-1} + b)
]

Here:

* (x_t) is the current input at time step (t)
* (h_{t-1}) is the hidden state from the previous time step
* (W_x) is the weight matrix applied to the input
* (W_h) is the weight matrix applied to the previous hidden state
* (b) is the bias term
* (\tanh) is the activation function that introduces non-linearity

This equation shows that the new hidden state is a **combination of the current input and the past memory**. Because the same weights (W_x) and (W_h) are reused at every time step, the network learns a consistent rule for updating its memory as it moves through the sequence.

The output of the network at each step (or sometimes only at the final step) is computed from the hidden state using another transformation:

[
y_t = W_y h_t + b_y
]

where (W_y) is the output weight matrix and (y_t) is the predicted output.

During the **forward pass**, the RNN processes the sequence step by step. Starting with an initial hidden state (usually a vector of zeros), the network reads the first input, updates the hidden state, then moves to the next input while carrying the updated hidden state forward. In this way, the hidden state acts as a **summary of all the inputs seen so far**.

For example, consider a temperature sequence:

```
Day1: 30
Day2: 31
Day3: 29
Day4: 28
```

The RNN processes it as follows:

```
Step1: h1 = f(x1, h0)
Step2: h2 = f(x2, h1)
Step3: h3 = f(x3, h2)
Step4: h4 = f(x4, h3)
```

After processing the sequence, the final hidden state (h_4) contains information about the entire sequence. This state can then be used to predict the next temperature.

During training, the network learns the weights (W_x), (W_h), and (W_y) using **Backpropagation Through Time (BPTT)**. In this process, the prediction error is propagated backward through all the time steps of the sequence, adjusting the weights to minimize the loss function. Because the same weights are reused at every step, updates affect how the network processes the entire sequence.

One limitation of standard RNNs is the **vanishing gradient problem**, where gradients become extremely small when sequences are long. This makes it difficult for the network to learn long-term dependencies. To address this, advanced architectures such as **LSTM (Long Short-Term Memory)** and **GRU (Gated Recurrent Units)** were developed, which include gating mechanisms that help preserve important information over longer sequences.

In summary, the working of an RNN involves **processing inputs sequentially, updating an internal hidden state that acts as memory, and using this memory to produce predictions**. By repeatedly combining the current input with information from previous steps, the RNN learns patterns across time and becomes capable of modeling sequential relationships in data such as text, speech, weather patterns, or financial time series.
