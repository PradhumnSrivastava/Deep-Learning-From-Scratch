# **Early Stopping**


## **Definition**

Early Stopping is a regularization technique used in deep learning to prevent overfitting. It stops the training process when the model’s performance on the validation dataset stops improving. Instead of training for a fixed number of epochs, the training process is automatically halted once the validation loss begins to increase or stops decreasing.

---

## **Key Idea**

During training:

- Training loss usually decreases continuously.
- Validation loss decreases initially but may increase after some time.

When validation loss starts increasing, it indicates that the model is beginning to **memorize the training data instead of learning general patterns**.

---

## **Stopping Condition**

```
Training Loss ↓
Validation Loss ↑
```

When this condition occurs consistently for several epochs, training is stopped.

---

## **Example Code**

```python
from tensorflow.keras.callbacks import EarlyStopping

early_stop = EarlyStopping(
    monitor='val_loss',
    patience=5,
    restore_best_weights=True
)
```

# Data Scaling

---

## **Definition**

Data Scaling is a preprocessing technique used to transform input features so that they have a **similar range or distribution**. Neural networks train more efficiently when input values are scaled because large differences in feature magnitudes can slow down the learning process.

---

## **Why Data Scaling is Important**

Neural networks rely on **gradient descent optimization**, and large feature values can cause unstable gradients. Scaling ensures that all features contribute equally during training.

---

## **Common Scaling Methods**

| Method | Description |
|------|------|
| **Min-Max Scaling** | Scales data to a fixed range (usually 0 to 1) |
| **Standardization** | Centers data around mean 0 with standard deviation 1 |
| **Robust Scaling** | Uses median and interquartile range |

---

## **Min-Max Scaling Formula**

```
X_scaled = (X - X_min) / (X_max - X_min)
```

---

## **Standardization Formula**

```
X_scaled = (X - μ) / σ
```

Where:

- μ = mean  
- σ = standard deviation  

---

## **Example Code**

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

# Dropout Layers

---

## **Definition**

Dropout is a regularization technique where **random neurons are temporarily deactivated during training**. This prevents the neural network from relying too heavily on specific neurons and encourages the network to learn more robust features.

---

## **How Dropout Works**

During training:

- A fraction of neurons is randomly removed.
- The network is forced to learn redundant representations.

During inference (testing):

- All neurons are active.

---

## **Dropout Concept**

```
Neuron Output → Randomly set to 0 with probability p
```

Where:

- p = dropout rate

Example:

```
Dropout Rate = 0.5
```

This means **50% of neurons are randomly disabled during training**.

---

## **Example Code**

```python
from tensorflow.keras.layers import Dropout

model.add(Dropout(0.5))
```

# Regularization

---

## **Definition**

Regularization is a technique used to **reduce model complexity and prevent overfitting** by adding a penalty to large model weights. This encourages the neural network to learn simpler and more generalizable patterns.

---

## **Types of Regularization**

| Type | Description |
|------|------|
| **L1 Regularization** | Adds absolute value of weights |
| **L2 Regularization** | Adds squared value of weights |
| **Elastic Net** | Combination of L1 and L2 |

---

## **L1 Regularization Formula**

```
Loss = Original Loss + λ Σ |w|
```

---

## **L2 Regularization Formula**

```
Loss = Original Loss + λ Σ w²
```

Where:

- λ = regularization strength  
- w = model weights  

---

## **Example Code**

```python
from tensorflow.keras.regularizers import l2

Dense(128, activation='relu', kernel_regularizer=l2(0.01))
```
