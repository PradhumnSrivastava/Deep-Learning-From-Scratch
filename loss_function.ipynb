# **Loss Function**

Loss functions are a fundamental component in machine learning, quantifying the difference between the predicted outputs of a model and the actual target values. They guide the optimization process by providing a measurable way to gauge the performance and accuracy of a model.

### **MSE**

Mean Squared Error (MSE) is a loss function used mainly for regression tasks. It calculates the average of the squared differences between the actual values and the predicted values. Squaring the error ensures that larger errors receive a higher penalty, which helps the model focus more on reducing large prediction mistakes.

MSE = (1/n) Σ (y - ŷ)²

```python
import numpy as np

def mean_squared_error(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)
```

### **Mean Absolute Error (MAE)**

Mean Absolute Error (MAE) measures the average magnitude of the errors between predicted values and actual values without considering their direction. It calculates the absolute difference between predicted and true values, making it less sensitive to outliers compared to Mean Squared Error.

MAE = (1/n) Σ |y - ŷ|

```python
def mean_absolute_error(y_true, y_pred):
    return np.mean(np.abs(y_true - y_pred))
```

### **Binary Cross Entropy**

Binary Cross Entropy (BCE) is a loss function used in binary classification problems. It measures the difference between the predicted probability distribution and the actual binary labels. The loss increases when the predicted probability diverges from the true label.

BCE = -(1/n) Σ [y log(ŷ) + (1-y) log(1-ŷ)]

```python
def binary_cross_entropy(y_true, y_pred):
    epsilon = 1e-15
    y_pred = np.clip(y_pred, epsilon, 1 - epsilon)
    return -np.mean(y_true*np.log(y_pred) + (1-y_true)*np.log(1-y_pred))
```

### **Categorical Cross Entropy**

Categorical Cross Entropy is used in multi-class classification problems where the model predicts probabilities for multiple classes. It measures how well the predicted probability distribution matches the true distribution represented by one-hot encoded labels.

CCE = - Σ y log(ŷ)

```python
def categorical_cross_entropy(y_true, y_pred):
    epsilon = 1e-15
    y_pred = np.clip(y_pred, epsilon, 1 - epsilon)
    return -np.sum(y_true * np.log(y_pred))
```

### **Hinge Loss** 

Hinge Loss is primarily used in Support Vector Machines and classification tasks. It penalizes predictions that fall on the wrong side of the decision boundary or within the margin, encouraging the model to create a larger margin between classes.

Loss = max(0, 1 - y * ŷ)

```python
def hinge_loss(y_true, y_pred):
    return np.mean(np.maximum(0, 1 - y_true * y_pred))
```

### **Huber Loss**

Huber Loss combines the advantages of Mean Squared Error and Mean Absolute Error. For small errors, it behaves like MSE, and for large errors it behaves like MAE. This property makes it more robust to outliers while still maintaining smooth gradients for optimization.

Lδ(y, ŷ) = { 
  1/2 (y − ŷ)²                if |y − ŷ| ≤ δ
  δ(|y − ŷ| − 1/2 δ)          if |y − ŷ| > δ
}

```python
def huber_loss(y_true, y_pred, delta=1.0):
    error = y_true - y_pred
    condition = np.abs(error) <= delta
    squared_loss = 0.5 * error**2
    linear_loss = delta * (np.abs(error) - 0.5 * delta)
    return np.mean(np.where(condition, squared_loss, linear_loss))
```
