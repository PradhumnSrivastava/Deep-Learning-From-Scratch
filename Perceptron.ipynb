# **Perceptron**

The perceptron is a fundamental algorithm used for binary classification in machine learning and neural networks. It calculates a weighted sum of input features and applies an activation function to produce the final output. The linear combination is written as:

z = w₁x₁ + w₂x₂ + ... + wₙxₙ + b

where w₁, w₂, …, wₙ represent the weights, x₁, x₂, …, xₙ represent the input features, and b is the bias term. The output of the perceptron is computed using an activation function:
y = step(z)

During training, the weights are updated using the rule:

w = w + η (y − ŷ) x

where η is the learning rate.

## **Mathematical Formula Used**

### Linear Combination

\[
z = w_1x_1 + w_2x_2 + ... + w_nx_n + b
\]

Where:

- \(w_1, w_2, ..., w_n\) = weights  
- \(x_1, x_2, ..., x_n\) = input features  
- \(b\) = bias  
- \(z\) = linear output

---

### Activation Function

\[
y = step(z)
\]

Where:

- \(z\) = linear combination output  
- \(y\) = predicted output after activation

---

### Weight Update Rule

\[
w = w + \eta (y - \hat{y})x
\]

Where:

- \(w\) = weight  
- \(\eta\) = learning rate  
- \(y\) = actual value (true label)  
- \(\hat{y}\) = predicted value  
- \(x\) = input feature

## **Import Required Libraries**

```python
import numpy as np
```

## **Step Function (Activation Function)**

```python
def step_function(x):
    if x >= 0:
        return 1
    else:
        return 0

#vectorize version of step function
def step_function(x):
    return np.where(x >= 0, 1, 0)
```

## **Basic Perceptron Model (From Scratch)**

```python
class Perceptron:

    def __init__(self, learning_rate=0.01, epochs=100):
        self.lr = learning_rate
        self.epochs = epochs

    def fit(self, X, y):
        n_samples, n_features = X.shape
        
        # initialize weights and bias
        self.weights = np.zeros(n_features)
        self.bias = 0

        for _ in range(self.epochs):
            for i in range(n_samples):

                linear_output = np.dot(X[i], self.weights) + self.bias
                y_pred = step_function(linear_output)

                update = self.lr * (y[i] - y_pred)

                self.weights += update * X[i]
                self.bias += update

    def predict(self, X):
        linear_output = np.dot(X, self.weights) + self.bias
        return step_function(linear_output)
```

## **Simple Dataset Example**

```python
X = np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])

y = np.array([0,0,0,1])
```

## **Train the Perceptron**

```python
model = Perceptron(learning_rate=0.1, epochs=10)
model.fit(X, y)
```

## **Make Predictions**

```python
predictions = model.predict(X)

print("Predictions:", predictions)
```

## **Print Weights and Bias**

```python
print("Weights:", model.weights)
print("Bias:", model.bias)
```

## **Accuracy Function**

```python
def accuracy(y_true, y_pred):
    return np.sum(y_true == y_pred) / len(y_true)


acc = accuracy(y, predictions)
print("Accuracy:", acc)
```

## **Full Minimal Perceptron Code**

```python
import numpy as np

X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([0,0,0,1])

weights = np.zeros(2)
bias = 0
lr = 0.1
epochs = 10

def step(x):
    return 1 if x >= 0 else 0

for _ in range(epochs):
    for i in range(len(X)):

        z = np.dot(X[i], weights) + bias
        y_pred = step(z)

        weights += lr * (y[i] - y_pred) * X[i]
        bias += lr * (y[i] - y_pred)

print("Weights:", weights)
print("Bias:", bias)
```

# **Multi Layer Perceptron**

## **Import Library**

```python
import numpy as np
```

## **Activation Function**

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))


#sigmoid derivative
def sigmoid_derivative(x):
    return x * (1 - x)
```

## **Dataset (XOR Problem)**

```python
X = np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])

y = np.array([
    [0],
    [1],
    [1],
    [0]
])
```

## **Initialize Weights and Bias**

```python
input_size = 2
hidden_size = 4
output_size = 1

weights_input_hidden = np.random.rand(input_size, hidden_size)
bias_hidden = np.random.rand(1, hidden_size)

weights_hidden_output = np.random.rand(hidden_size, output_size)
bias_output = np.random.rand(1, output_size)
```

## **Training the Network**

```python
learning_rate = 0.1
epochs = 10000

for epoch in range(epochs):

    # Forward propagation
    hidden_input = np.dot(X, weights_input_hidden) + bias_hidden
    hidden_output = sigmoid(hidden_input)

    final_input = np.dot(hidden_output, weights_hidden_output) + bias_output
    predicted_output = sigmoid(final_input)

    # Error
    error = y - predicted_output

    # Backpropagation
    d_predicted = error * sigmoid_derivative(predicted_output)

    hidden_error = d_predicted.dot(weights_hidden_output.T)
    d_hidden = hidden_error * sigmoid_derivative(hidden_output)

    # Update weights
    weights_hidden_output += hidden_output.T.dot(d_predicted) * learning_rate
    bias_output += np.sum(d_predicted, axis=0, keepdims=True) * learning_rate

    weights_input_hidden += X.T.dot(d_hidden) * learning_rate
    bias_hidden += np.sum(d_hidden, axis=0, keepdims=True) * learning_rate
```
